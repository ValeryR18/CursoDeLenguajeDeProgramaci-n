```python
# Diccionario para almacenar los datos de los clientes
portafolio = {}

# Función para ingresar un cliente
def ingresar_cliente():
    clear_output()  # Limpia la salida

    # Widgets para ingresar datos
    cliente = widgets.Text(description='Cliente', placeholder='Ingresa el Nombre')
    monto_invertido = widgets.FloatText(description='Monto Invertido', value=0.0)
    rendimiento_anual = widgets.FloatText(description='Rendimiento Anual (%)', value=0.0)

    # Botón para guardar cliente
    button = widgets.Button(description="Guardar")
    output = widgets.Output()

    def on_button_clicked(b):
        # Guardar los datos en el diccionario
        if cliente.value:
            portafolio[cliente.value] = {
                "Monto Invertido": monto_invertido.value,
                "Rendimiento Anual": rendimiento_anual.value
            }
            with output:
                clear_output()
                print(f"Cliente {cliente.value} guardado en el portafolio.")
        else:
            with output:
                clear_output()
                print("Por favor, ingresa un nombre de cliente.")

    button.on_click(on_button_clicked)

    display(cliente, monto_invertido, rendimiento_anual, button, output)

# Ejecutar la función para ingresar cliente
ingresar_cliente()
```

# **Valor Presente Neto (VPN)**

VPN.png

## **VPN de cada proyecto de flujo de caja**

```python
df.columns = df.columns.str.replace('–', '_')  # Reemplaza guion largo Unicode por guion bajo
df.columns = df.columns.str.replace('A\x96o_', 'Año_', regex=False)       # Elimina espacios adicionales
```

```python
print(df.columns.tolist())
```

```python
#VPN de cada proyecto
# Convertir columnas numéricas a tipo float (eliminar caracteres no numéricos si los hay)
for col in df.columns[1:]:
    df[col] = df[col].replace('[^\d\.-]', '', regex=True).astype(float)

# Tasa de descuento fija
r = 0.10

# Función para calcular VPN usando la fórmula tradicional
def calcular_vpn(row, tasa):
    vpn = 0
    num_flujos = len([col for col in df.columns if col.startswith('Año_')])
    for t in range(0, num_flujos + 1):  # Incluye t = 0
        if t == 0:
            flujo = row['Inversion_Inicial']
        else:
            col_name = f'Año_{t}'
            flujo = row[col_name] if col_name in row else 0
        valor_presente = flujo / ((1 + tasa) ** t)
        vpn += valor_presente
    return vpn

# Aplicar cálculo a cada fila
df['VPN'] = df.apply(lambda row: calcular_vpn(row, r), axis=1)

# Mostrar resultados VPN de cada proyecto
print(df[['Proyecto', 'VPN']])
```

## **Tasa Interna de Retorno TIR**

```python
#Se instala la libreria Numpy financial
!pip install numpy-financial
```

```python
import numpy_financial as npf

# Función para obtener lista de flujos de caja por proyecto
def obtener_flujos(row):
    flujos = [row['Inversion_Inicial']]  # Año 0
    for t in range(1, 11):  # Años 1 a 10
        col_name = f'Año_{t}'
        flujos.append(row[col_name] if col_name in row else 0)
    return flujos

# Función para calcular TIR
def calcular_tir(row):
    flujos = obtener_flujos(row)
    return npf.irr(flujos)

# Aplicar TIR a cada fila del DataFrame
df['TIR'] = df.apply(calcular_tir, axis=1)

# Mostrar resultados
print(df[['Proyecto', 'TIR']])
```

## **Gráfico de la TIR**

```python
# 4. GRAFICA CON MATPLOTLIB DE TIR
import matplotlib.ticker as mtick

# Ordenar por TIR para una mejor visualización (opcional, pero recomendado)
df_sorted_tir = df.sort_values(by='TIR', ascending=False)

# Crear la figura y los ejes para el gráfico
plt.figure(figsize=(15, 9)) # Ajustar tamaño para mejor legibilidad con muchos proyectos

# Crear el gráfico de barras
# Usamos un color diferente para distinguir del gráfico de VPN si se generasen juntos
bars_tir = plt.bar(df_sorted_tir['Proyecto'], df_sorted_tir['TIR'], color='mediumseagreen')

# Añadir títulos y etiquetas descriptivas
plt.title('Tasa Interna de Retorno (TIR) por Proyecto', fontsize=18, pad=20)
plt.xlabel('Proyecto', fontsize=14, labelpad=15)
plt.ylabel('TIR (%)', fontsize=14, labelpad=15)

# Formatear el eje Y para mostrar los valores como porcentajes
# xmax=1.0 indica que un valor de 1.0 en los datos representa 100%
# decimals=1 muestra un decimal en el porcentaje (ej. 15.5%)
plt.gca().yaxis.set_major_formatter(mtick.PercentFormatter(xmax=1.0, decimals=1))

# Rotar las etiquetas del eje X para que no se solapen y mejorar legibilidad
plt.xticks(rotation=70, ha="right", fontsize=10) # 'ha' es horizontal alignment
plt.yticks(fontsize=10)

# Añadir una cuadrícula horizontal para facilitar la lectura de los valores de la TIR
plt.grid(True, linestyle='--', alpha=0.6, axis='y') # 'axis='y'' para solo líneas horizontales

# Mostrar los valores de la TIR encima de cada barra
for bar in bars_tir:
    yval = bar.get_height()
    # Formatear el número como porcentaje con un decimal
    # Ajustar la posición vertical del texto (va)
    # y la posición horizontal (ha)
    plt.text(
        bar.get_x() + bar.get_width() / 2.0,
        yval,
        f'{yval:.1%}', # Formato de porcentaje con 1 decimal
        va='bottom' if yval >= 0 else 'top', # Colocar encima para positivos, debajo para negativos
        ha='center',
        fontsize=8,
        color='black' # Color del texto de la etiqueta
    )

# Ajustar el layout para que todos los elementos (títulos, etiquetas) quepan bien
plt.tight_layout(pad=1.5) # Añadir un poco de padding

# Mostrar el gráfico
plt.show()
```
